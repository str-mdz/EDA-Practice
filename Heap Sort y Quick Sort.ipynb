**Quick sort**

```python
#Código funcional 

def partition(arr, low, high):
    i = low-1
    pivot = arr[high]
    for j in range(low , high):
        if   arr[j] <= pivot:
            i = i+1
            arr[i],arr[j] = arr[j],arr[i]
 
    arr[i+1],arr[high] = arr[high],arr[i+1]
    return i+1


def quick_sort(arr, low, high):
    if low < high:
        pivot_Index = partition(arr, low, high)
        quick_sort(arr, low, pivot_Index-1)
        quick_sort(arr, pivot_Index+1, high)
    return arr

arr = []
for i in range(10):
    arr.append(random.randint(0,20))
    
print("Lista original: ", arr)
print("Lista ordenada: ", quick_sort(arr,0,len(arr)-1))
```

## Mejor caso

```python

def partitionMC(arr, low, high):
    global timeMC
    i = low-1                                               # 5
    pivot = arr[(low + high)//2]                            # 8
    j = low                                                 # 3

    while j < high:                                         # 4n + 4
        timeMC += 1
        if arr[j] <= pivot:                                 # 5n
            i = i+1                                         # 9n
            arr[i],arr[j] = arr[j],arr[i]                   # 5n
        j += 1                                              # 5n
    arr[i+1],arr[high] = arr[high],arr[i+1]                 # 13
    return i+1                                              # 4
                                                            # 28n + 37

def quick_sortMC(arr, low, high):
    global timeMC
    if low < high:                                          # 4 * (log2(n)
        timeMC += 1
        pivot_Index = partitionMC(arr, low, high)           # 6 * (28n + 37) * (log2(n))
        quick_sortMC(arr, low, pivot_Index-1)               # 6 * (log2(n)) 
        quick_sortMC(arr, pivot_Index+1, high)              # 6 * (log2(n))
    return arr                                              # 2 * (log2(n)
                                                            # 168nlog2(n) + 240log2(n)
```
