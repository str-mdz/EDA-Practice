```python
import matplotlib.pyplot as plt
import random
```

**Bubble Sort**

```python
#Implementación de Bubble Sort con contador time
def bubble_sort(arr):
    time = 0                                                  
    n = len(arr)                                            # 4                                                                                                                                                                        
    i=0                                                     # 3
    while i < n:                                            # 4(n + 1) = 4n + 4
        time += 1                                           
        j = 0                                               # 3n
        while j < n-i-1:                                    # 8n (n + 1)/2 = (8n^2 + 8n)/2 = 4n^2 + 4n
            time += 1                                       
            if (arr[j] > arr[j+1]):                         # 8 (n^2)
                arr[j], arr[j+1] = arr[j+1], arr[j]         # 13 (n^2)
            j+=1                                            # 5 (n^2)
        i+=1                                                # 5n
                                                            # 30n^2 + 16n + 11
    return time
```

- **Polinomio:** 30n^2 + 16n + 11
- **Complejidad mejor caso:** O(n^2)
- **Complejidad peor caso:** O(n^2)
- **Complejidad caso promedio:** O(n^2)

```python
#Lista valores random

x = []
y = []
i = 0
listaRandom = []

while i < 100:
    listaRandom.append(random.randint(-10, 10000))
    i += 1
    #print(listaRandom)
    x.append(len(listaRandom))
    y.append(bubble_sort(listaRandom))
    
plt.plot(x, y, "r*")
plt.title("Bubble Sort: Caso promedio")
plt.xlabel("Instancias de entrada")
plt.ylabel("Tiempo de ejecución")
plt.grid()
plt.show()
```

```python
#Lista valores random ordenados
listaOrdenada = []
i = 0
while i < 100:
    listaOrdenada.append(random.randint(-20, 20))
    i += 1
    listaOrdenada.sort()
    #print(listaOrdenada)
    x.append(len(listaOrdenada))
    y.append(bubble_sort(listaOrdenada))
    
plt.plot(x, y, "r*")
plt.title("Bubble Sort: Mejor caso")
plt.xlabel("Instancias de entrada")
plt.ylabel("Tiempo de ejecución")
plt.grid()
plt.show()
 
```

**Merge Sort**

```python
def merge_sort(arr):
    global time
    if len(arr) <= 1:                               # 5 * log2(n)
        time += 1
        return arr

    mid = len(arr) // 2                             # 6 * log2(n)
    left = arr[:mid]                                # 4 * log2(n)
    right = arr[mid:]                               # 4 * log2(n)

    left = merge_sort(left)                         # 4 * log2(n) * log2(n)
    right = merge_sort(right)                       # 4 * log2(n) * log2(n)

    return merge(left, right)                       # (34n + 27)(log2(n)) = 34n * log2(n) + 27 log2(n)

def merge(left, right):
    global time
    resultado = []                                  # 3
    i = j = 0                                       # 6
    
    while i < len(left) and j < len(right):         # 10 * (n + 1) = 10n + 10
        if left[i] < right[j]:                      # 6n
            time += 1
            resultado.append(left[i])               # 4n
            i += 1                                  # 5n
        else:                                           
            time += 1
            resultado.append(right[j])              # 4n
            j += 1                                  # 5n
            
    resultado.extend(left[i:])                      # 4
    resultado.extend(right[j:])                     # 4

    return resultado
                                                    # 34n + 27
                                            
#Complejidad total = 34n*log2(n) + 8log2(n)^2 + 50log2(n)

```

- **Polinomio:** 34n*log2(n) + 8log2(n)^2 + 50log2(n)
- **Complejidad mejor caso:** O(n*log2(n))
- **Complejidad peor caso:** O(n*log2(n))
- **Complejidad caso promedio:** O(n*log2(n))

```python
#Lista valores random

x = []
y = []
i = 0
listaRandom = []


while i < 100:
    time = 0
    j = 0
    listaRandom = []
    while j < i:
        listaRandom.append(random.randint(-100, 100))
        j += 1
    i += 1
    x.append(len(listaRandom))
    #print(listaRandom)
    listaRandom = merge_sort(listaRandom)
    #print(listaRandom)
    y.append(time)
    
plt.plot(x, y, "r*")
plt.title("Merge Sort: Caso promedio")
plt.xlabel("Instancias de entrada")
plt.ylabel("Tiempo de ejecución")
plt.grid()
plt.show()
```
