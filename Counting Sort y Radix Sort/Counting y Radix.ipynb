## Counting sort

C Ó D I G O

```python
def countingSort(array):
    size = len(array)
    output = [0] * size
    maximo = max(array)+1
    # Inicializa el arreglo de conteo
    count = [0] * maximo

    # Almacena el conteo de cada elemento en el arreglo count
    i = 0
    while i < size:
        count[array[i]] += 1
        i += 1

    # Almacena el conteo acumulativo
    i = 1
    while i < maximo:
        count[i] += count[i - 1]
        i += 1

    # Encuentra el índice de cada elemento del arreglo original en el arreglo count
    # y coloca los elementos en el arreglo de salida
    i = size - 1
    while i >= 0:
        output[count[array[i]] - 1] = array[i]
        count[array[i]] -= 1
        i -= 1

    # Copia los elementos ordenados al arreglo original
    i = 0
    while i < size:  
        array[i] = output[i]
        i += 1


data = [random.randint(0, 50) for i in range(0, 11)]
print(data)
countingSort(data)
print("Sorted Array in Ascending Order: ")
print(data)
```

```python
def countingSortID(array):
    time = 0
    space = 0 
    
    size = len(array)                                   # 4 n
    output = [0] * size                                 # 5
    space += size
    maximo = maxi(array)+1                              # 8n
    count = [0] * maximo                                # 5
    space += maximo                                     
    
    i = 0                                               # 3
    while i < size:                                     # 4 (n + 1)
        time += 1
        count[array[i].id] += 1                         # 7 n
        i += 1                                          # 5 n

    i = 1                                               # 3
    while i < maximo:                                   # 4 (m + 1)
        time += 1
        count[i] += count[i - 1]                        # 7 m
        i += 1                                          # 5 m

    i = size - 1                                        # 5
    while i >= 0:                                       # 4 (n + 1)
        time += 1
        output[count[array[i].id] - 1] = array[i]       # 9 n
        count[array[i].id] -= 1                         # 7 n
        i -= 1                                          # 5 n

    i = 0                                               # 3
    while i < size:                                     # 4 (n + 1)
        time += 1
        array[i] = output[i]                            # 5 n
        i += 1                                          # 5 n
                                                        # 67n + 16m + 35
    return time, space

#countingSortID(listaNodos)
#for i in listaNodos:
    #print(str(i))
```

## RADIX SORT

```python
def countingSort(arr, exp1):

    n = len(arr)                                        
    output = [0] * n                                    

    count = [0] * 10                                    

    i = 0                                               
    while i < n:                                       
        index = arr[i] // exp1                          
        count[index % 10] += 1                         
        i += 1                                          

    i = 1                                               
    while i < 10:                                       
        count[i] += count[i - 1]                        
        i += 1                                          

    i = n - 1                                           
    while i >= 0:                                       
        index = arr[i] // exp1                          
        output[count[index % 10] - 1] = arr[i]          
        count[index % 10] -= 1                          
        i -= 1                                          

    i = 0                                               
    while i < n:                                        
        arr[i] = output[i]                              
        i += 1                                          

def radixSort(arr):
    max1 = max(arr)                                     
    exp = 1                                             
    while max1 // exp > 0:                              
        countingSort(arr, exp)                          
        exp *= 10                                       

# Código principal
arr = [170, 45, 75, 90, 802, 24, 2, 66]
radixSort(arr)

i = 0
while i < len(arr):
    print(arr[i], end=" ")
    i += 1
```
