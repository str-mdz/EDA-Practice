```python
def binary_search_graph_strings(l, k):
    time = 0
    i = 0
    f = len(l) - 1                             
    
    while (i <= f):
        time += 1
        m = (f + i)//2
        if (l[m].string == k):
            return time
        
        if (l[m].string < k):
            i = m+1
        else:
            f = m-1
    return time
```

```python
def lineal_recursiva_num(arr,val,i):
    global time
    time += 1
    if (len(arr)-1 < i):               
        return time                             
    
    if(arr[i].num == val):                  
        return time                            
    
    return lineal_recursiva_num(arr,val,i+1)
```
